```julia
using DifferentialEquations, Plots
```

```julia
function BelZha(du, u, p, t)
    x = u[1]
    y = u[2]
    z = u[3]
    
    s = p[1]
    w = p[2]
    q = p[3]
    
    du[1] = s * (y + x*(1 - q*x - y))  # dx/dt
    du[2] = (-y -x*y + z) / s  # dy/dt
    du[3] = w * (x - z)  # dz/dt
    
end

p = [77.27, 0.161, 8.375 * 10e-6]
u0 = [1., 2., 3.]
tspan = (0.0, 360.0)
prob = ODEProblem(BelZha, u0, tspan, p)
sol = solve(prob)
plot(sol)
```
